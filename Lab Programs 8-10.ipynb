```python
import pandas as pd

data = {
    'Product_Name': ['Laptop', 'Mobile', 'Laptop', 'Keyboard', 'Mobile', 'Mouse', 'Headphone'],
    'Price': [50000, 30000, 55000, 2000, 35000, 1000, 5000]
}

df = pd.DataFrame(data)

top_5_products = df.groupby("Product_Name")["Price"].sum().nlargest(5)

print("Five most sold products:")
print(top_5_products)
```

```python
import pandas as pd

data = {
    'Location': ['Chennai', 'Chennai', 'Coimbatore', 'Madurai', 'Coimbatore'],
    'beds': [3, 5, 4, 2, 5],
    'price': [6500000, 9500000, 5500000, 4200000, 7500000],
    'size': [1500, 3000, 2000, 1200, 3500]
}

df = pd.DataFrame(data)

location_price_average = df.groupby("Location")["price"].mean()

more_than_4beds = df[df["beds"] > 4].count()["beds"]

largest_sq_feet = df["size"].max()

print("Average house price for each location:")
print(location_price_average)

print("\nNumber of houses with more than 4 bedrooms:", more_than_4beds)

print("\nLargest house size:", largest_sq_feet)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month_sales': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Price': [50000, 45000, 60000, 55000, 70000, 65000]
}

df = pd.DataFrame(data)

monthly_sales = df.groupby("Month_sales")["Price"].sum()

# Line Graph
plt.figure(figsize=(8,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar Graph
plt.figure(figsize=(8,5))
plt.bar(monthly_sales.index, monthly_sales.values)
plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
