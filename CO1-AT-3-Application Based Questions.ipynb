```python
import numpy as np

# House data: Bedrooms, Square Footage, Sale Price
house_data = np.array([
    [3, 1400, 7500000],
    [5, 2200, 12500000],
    [4, 1800, 9800000],
    [6, 3000, 18500000],
    [2, 1100, 5200000]
])


filtered_houses = house_data[house_data[:, 0] > 4]


average_sale_price = np.mean(filtered_houses[:, 2])


print("Average Sale Price of houses with more than 4 bedrooms:", average_sale_price)
```

```python
from sklearn.linear_model import LogisticRegression


X = [
    [1, 3, 120],
    [6, 15, 300],
    [0, 1, 80],
    [4, 10, 250],
    [2, 5, 160]
]


y = [0, 1, 0, 1, 0]


model = LogisticRegression(max_iter=1000)
model.fit(X, y)


new_email = [[6, 15, 300]]

prediction = model.predict(new_email)

if prediction[0] == 1:
    print("Prediction: Spam Email")
else:
    print("Prediction: Not Spam")
```

```python
from sklearn.linear_model import LogisticRegression

# Training data
X = [
    [1, 3, 120],
    [6, 15, 300],
    [0, 1, 80],
    [4, 10, 250],
    [2, 5, 160]
]

# Target values
y = [0, 1, 0, 1, 0]

# Train the model
model = LogisticRegression(max_iter=1000)
model.fit(X, y)

# New email
new_email = [[5, 12, 280]]

# Prediction
prediction = model.predict(new_email)

if prediction[0] == 1:
    print("Prediction: Spam Email")
else:
    print("Prediction: Not Spam")
```

```python
from sklearn.linear_model import LinearRegression

# Features: Age, Kms_Driven, Engine_CC
X = [
    [2, 25000, 1200],
    [5, 60000, 1500],
    [3, 40000, 1300],
    [7, 90000, 1800],
    [1, 15000, 1000]
]

# Target: Resale Price
y = [650000, 420000, 520000, 300000, 720000]

# Train model
model = LinearRegression()
model.fit(X, y)

# User input
age = int(input("Enter Car Age (Years): "))
kms = int(input("Enter Kilometers Driven: "))
engine = int(input("Enter Engine Capacity (CC): "))

# Predict
new_car = [[age, kms, engine]]
predicted_price = model.predict(new_car)

print("Estimated Resale Price: ₹", round(predicted_price[0], 2))
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text

# Features: Income, Credit Score, Existing Loans
X = [
    [45000, 720, 0],
    [30000, 650, 1],
    [60000, 780, 0],
    [28000, 620, 2],
    [52000, 710, 1]
]

# Target: Approved
y = [1, 0, 1, 0, 1]

# Train CART model
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# User input
income = int(input("Enter Monthly Income: "))
credit = int(input("Enter Credit Score: "))
loans = int(input("Enter Existing Loans: "))

# Predict
new_applicant = [[income, credit, loans]]
prediction = model.predict(new_applicant)

if prediction[0] == 1:
    print("Loan Approved")
else:
    print("Loan Rejected")

# Display Decision Rules
print("\nDecision Rules:\n")
print(export_text(model,
                  feature_names=["Income", "Credit_Score", "Existing_Loans"]))
```
